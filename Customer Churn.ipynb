As we can see, TotalCharges column is int but was written as object so we have to convert it to int

```python
df.TotalCharges=pd.to_numeric(df["TotalCharges"],errors="coerce")
df.info()
```

We can see that its non-null values have gone from 7043 to 7032 indicating 11 missing values.11 is 0.15% of the entire dataset, we could impute or we could just drop the rows.I am dropping because 0.15% is little

```python
df.dropna(inplace=True)
df.info()
```

```python
df.tenure
```

We should use feature binning here to group this

```python
labels=["{0}-{1}".format(i,i+11) for i in range(1,72,12)]
labels
df.tenure_group=pd.cut(df["tenure"],range(1,80,12),right=False,labels=labels)
df.tenure_group
```
