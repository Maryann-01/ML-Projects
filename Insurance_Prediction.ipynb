```python
import numpy as numpy
import pandas as pd
import matplotlib.pyplot as plt
```

```python
train_df=pd.read_csv("/content/train_data.csv")
test_df=pd.read_csv("/content/test_data.csv")
train_df.info()
```

```python
train_df.YearOfObservation.value_counts()
len(train_df.YearOfObservation.unique())
len(train_df.Residential.unique())
```

```python
len(train_df.Insured_Period.unique())  #remove this, it has alot of unique values
len(train_df["Building Dimension"].unique())   #2044 unique values
len(train_df["Geo_Code"].unique())  #1308 unique values
train_df.Geo_Code.value_counts()
train_df["NumberOfWindows"].value_counts()
```

```python
train_df.drop(columns=["Customer Id"],inplace=True)
test_ids = test_df["Customer Id"]
test_df.drop(columns=["Customer Id"],inplace=True)
```

To handle missing values now

```python
train_df.info()
```

```python
test_df.info()
```

```python
cat_cols_train=train_df.select_dtypes(include=["object"]).columns.tolist()
cat_cols_test=test_df.select_dtypes(include=["object"]).columns.tolist()
num_cols_train=train_df.select_dtypes(include=["int64","float64"]).columns.tolist()
num_cols_test=test_df.select_dtypes(include=["int64","float64"]).columns.tolist()
cat_cols_train
```

To impute missing values

```python
train_df[num_cols_train]=train_df[num_cols_train].fillna(train_df[num_cols_train].median())
test_df[num_cols_test]=test_df[num_cols_test].fillna(test_df[num_cols_test].median())
train_df[cat_cols_train]=train_df[cat_cols_train].fillna(train_df[cat_cols_train].mode().iloc[0])
test_df[cat_cols_test]=test_df[cat_cols_test].fillna(test_df[cat_cols_test].mode().iloc[0])
train_df.info()
# test_df[cat_cols_test]
```

Next up, scaling..if necessary

```python
train_df.describe()
```

Before we scale, we need to convert our object columns to int using one hot encoding

```python

# train_df=pd.get_dummies(train_df,columns=cat_cols_train, drop_first=True)
# test_df=pd.get_dummies(test_df,columns=cat_cols_test, drop_first=True)
# test_df

print("Before Encoding:", train_df.dtypes)
train_df = pd.get_dummies(train_df, columns=cat_cols_train, drop_first=True, dtype=int)
test_df=pd.get_dummies(test_df,columns=cat_cols_test, drop_first=True,dtype=int)
```

```python
# test_df.head()
train_df.head()
```

Now , we scale seeing how the std looks

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
# scaler.fit_transform()
# if "Claim" in train_df.columns:
#     print("Target column is present.")
# else:
#     print("Target column is missing.")

X_train=train_df.drop(columns=["Claim"])
y_train=train_df["Claim"]


X_test = test_df.reindex(columns=X_train.columns, fill_value=0)
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
X_test_scaled
```

Light boosting model

```python
import lightgbm as lgb
from sklearn.metrics import roc_auc_score


model = lgb.LGBMClassifier(n_estimators=200, learning_rate=0.05, random_state=42)
model.fit(X_train_scaled, y_train)


y_pred_proba = model.predict_proba(X_test_scaled)[:, 1]

```

```python
submission = pd.DataFrame({"Customer Id": test_ids, "Claim": y_pred_proba})


submission.to_csv("submission.csv", index=False)

print("Submission file ready for upload!")
```
